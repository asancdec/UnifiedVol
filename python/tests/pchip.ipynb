```python
import numpy as np

# Shape: len(tenors_lv) rows × len(spots) columns
local_vol = [
    [1.276298048063844, 1.003668468561284, 0.763341610134954, 0.656259555204840,
     0.558742412114170, 0.471309770180474, 0.394282909978121, 0.327851917007969,
     0.272248655323075, 0.227928097882543, 0.195651582913945, 0.176395090331952,
     0.170997369307363, 0.179338455273741, 0.198851733403139, 0.260693990665305,
     0.526409951630718],

    [1.109958678329192, 0.880069540214445, 0.674325251979984, 0.582983846356871,
     0.500159660821145, 0.426095298991793, 0.360787227785619, 0.304093415246649,
     0.255856470003856, 0.216007724440915, 0.184630914170353, 0.161985728378576,
     0.148506178827290, 0.144771664461366, 0.151332818588844, 0.189438124290329,
     0.179308989864150],

    [0.934585870404535, 0.751335542951122, 0.587389377652960, 0.513886691004124,
     0.446522127644016, 0.385522483561629, 0.331015803878904, 0.283064042306597,
     0.241705179415223, 0.206999019494865, 0.179073571676958, 0.158171771562970,
     0.144697387732397, 0.139238202190609, 0.142435418365131, 0.169266175997874,
     0.081995839668887],

    [0.840458432751015, 0.663873169348955, 0.516110831224588, 0.452649512309452,
     0.395794640161600, 0.345231244933347, 0.300632279215731, 0.261692899644489,
     0.228163245326485, 0.199876782361340, 0.176771953335320, 0.158903614083491,
     0.146434455563934, 0.139575780934685, 0.138381477318108, 0.147087321448847,
     0.069447471520729],

    [0.739172310199885, 0.588049001787340, 0.462097203208204, 0.407992636572875,
     0.359417584194405, 0.316061365643860, 0.277622772325442, 0.243831738691879,
     0.214467277624531, 0.189370185580346, 0.168449335267414, 0.151679569790949,
     0.139085876612592, 0.130698880828362, 0.126440051173445, 0.127081725025745,
     0.092384355693564],

    [0.636813930664798, 0.511353118288399, 0.409409008317520, 0.365708208007046,
     0.326206923765417, 0.290541500031864, 0.258442418740812, 0.229725174439492,
     0.204278718452905, 0.182052995574815, 0.163045722965612, 0.147286631346116,
     0.134814089487372, 0.125632000737889, 0.119617690875147, 0.114326022047595,
     0.086308232836129],

    [0.609013555458939, 0.450709739880755, 0.348912202681917, 0.312834716362299,
     0.283335922301813, 0.258448868456294, 0.236758377834133, 0.217347527935057,
     0.199676842664837, 0.183459409650013, 0.168558710497155, 0.154910961020767,
     0.142462551209494, 0.131108199149384, 0.120608525684912, 0.099497920755146,
     0.057155046544458],

    [0.455934760619074, 0.389483342800450, 0.314984746819418, 0.286493270284568,
     0.263109835893457, 0.243501558805801, 0.226506351242242, 0.211284455823126,
     0.197291501857117, 0.184196899114101, 0.171803076323602, 0.159982055452486,
     0.148628376510911, 0.137620776371215, 0.126781218735938, 0.104165989905991,
     0.072558104154688],

    [0.338762779671638, 0.327224125258228, 0.293924122433950, 0.275969638523198,
     0.258318129197125, 0.241350612335169, 0.225265025009697, 0.210156031216685,
     0.196059056699353, 0.182975413691723, 0.170886917929039, 0.159764455685317,
     0.149572973149845, 0.140274297671298, 0.131828600316140, 0.117331187097803,
     0.105733162148777],

    [0.345562397463100, 0.312034819477716, 0.278121561257451, 0.262323460332735,
     0.247485084412481, 0.233618472191296, 0.220687961302520, 0.208633502997005,
     0.197384108349205, 0.186865517868938, 0.177004431495477, 0.167730670009323,
     0.158978085231355, 0.150684708477313, 0.142792426692114, 0.127993944595741,
     0.114165094829937],

    [0.337647545911004, 0.296193296989319, 0.263259265271726, 0.249171678218518,
     0.236423000883613, 0.224882479348714, 0.214443137565403, 0.205013218785814,
     0.196511304123318, 0.188863539392619, 0.182002055945909, 0.175864058239769,
     0.170391277525532, 0.165529622455798, 0.161228932702541, 0.154128319007639,
     0.148759871661047]
]

moneyness_log = np.array([
    -0.6931472, -0.5108256, -0.3566749, -0.2876821, -0.2231436, -0.1625189,
    -0.1053605, -0.0512933,  0.0000000,  0.0487902,  0.0953102,  0.1397619,
     0.1823216,  0.2231436,  0.2623643,  0.3364722,  0.4054651
])

tenors = np.array([
    0.083333, 0.166667, 0.250000, 0.333333, 0.416667, 0.500000,
    0.750000, 1.000000, 1.500000, 2.000000, 3.000000
])
import numpy as np

# -------------------------------
# Data: local volatility surface
# -------------------------------
local_vol = [
    [1.276298048063844, 1.003668468561284, 0.763341610134954, 0.656259555204840,
     0.558742412114170, 0.471309770180474, 0.394282909978121, 0.327851917007969,
     0.272248655323075, 0.227928097882543, 0.195651582913945, 0.176395090331952,
     0.170997369307363, 0.179338455273741, 0.198851733403139, 0.260693990665305,
     0.526409951630718],

    [1.109958678329192, 0.880069540214445, 0.674325251979984, 0.582983846356871,
     0.500159660821145, 0.426095298991793, 0.360787227785619, 0.304093415246649,
     0.255856470003856, 0.216007724440915, 0.184630914170353, 0.161985728378576,
     0.148506178827290, 0.144771664461366, 0.151332818588844, 0.189438124290329,
     0.179308989864150],

    [0.934585870404535, 0.751335542951122, 0.587389377652960, 0.513886691004124,
     0.446522127644016, 0.385522483561629, 0.331015803878904, 0.283064042306597,
     0.241705179415223, 0.206999019494865, 0.179073571676958, 0.158171771562970,
     0.144697387732397, 0.139238202190609, 0.142435418365131, 0.169266175997874,
     0.081995839668887],

    [0.840458432751015, 0.663873169348955, 0.516110831224588, 0.452649512309452,
     0.395794640161600, 0.345231244933347, 0.300632279215731, 0.261692899644489,
     0.228163245326485, 0.199876782361340, 0.176771953335320, 0.158903614083491,
     0.146434455563934, 0.139575780934685, 0.138381477318108, 0.147087321448847,
     0.069447471520729],

    [0.739172310199885, 0.588049001787340, 0.462097203208204, 0.407992636572875,
     0.359417584194405, 0.316061365643860, 0.277622772325442, 0.243831738691879,
     0.214467277624531, 0.189370185580346, 0.168449335267414, 0.151679569790949,
     0.139085876612592, 0.130698880828362, 0.126440051173445, 0.127081725025745,
     0.092384355693564],

    [0.636813930664798, 0.511353118288399, 0.409409008317520, 0.365708208007046,
     0.326206923765417, 0.290541500031864, 0.258442418740812, 0.229725174439492,
     0.204278718452905, 0.182052995574815, 0.163045722965612, 0.147286631346116,
     0.134814089487372, 0.125632000737889, 0.119617690875147, 0.114326022047595,
     0.086308232836129],

    [0.609013555458939, 0.450709739880755, 0.348912202681917, 0.312834716362299,
     0.283335922301813, 0.258448868456294, 0.236758377834133, 0.217347527935057,
     0.199676842664837, 0.183459409650013, 0.168558710497155, 0.154910961020767,
     0.142462551209494, 0.131108199149384, 0.120608525684912, 0.099497920755146,
     0.057155046544458],

    [0.455934760619074, 0.389483342800450, 0.314984746819418, 0.286493270284568,
     0.263109835893457, 0.243501558805801, 0.226506351242242, 0.211284455823126,
     0.197291501857117, 0.184196899114101, 0.171803076323602, 0.159982055452486,
     0.148628376510911, 0.137620776371215, 0.126781218735938, 0.104165989905991,
     0.072558104154688],

    [0.338762779671638, 0.327224125258228, 0.293924122433950, 0.275969638523198,
     0.258318129197125, 0.241350612335169, 0.225265025009697, 0.210156031216685,
     0.196059056699353, 0.182975413691723, 0.170886917929039, 0.159764455685317,
     0.149572973149845, 0.140274297671298, 0.131828600316140, 0.117331187097803,
     0.105733162148777],

    [0.345562397463100, 0.312034819477716, 0.278121561257451, 0.262323460332735,
     0.247485084412481, 0.233618472191296, 0.220687961302520, 0.208633502997005,
     0.197384108349205, 0.186865517868938, 0.177004431495477, 0.167730670009323,
     0.158978085231355, 0.150684708477313, 0.142792426692114, 0.127993944595741,
     0.114165094829937],

    [0.337647545911004, 0.296193296989319, 0.263259265271726, 0.249171678218518,
     0.236423000883613, 0.224882479348714, 0.214443137565403, 0.205013218785814,
     0.196511304123318, 0.188863539392619, 0.182002055945909, 0.175864058239769,
     0.170391277525532, 0.165529622455798, 0.161228932702541, 0.154128319007639,
     0.148759871661047]
]

local_vol_np = np.array(local_vol).T

moneyness_log = np.array([
    -0.6931472, -0.5108256, -0.3566749, -0.2876821, -0.2231436, -0.1625189,
    -0.1053605, -0.0512933,  0.0000000,  0.0487902,  0.0953102,  0.1397619,
     0.1823216,  0.2231436,  0.2623643,  0.3364722,  0.4054651
])

tenors = np.array([
    0.083333, 0.166667, 0.250000, 0.333333, 0.416667, 0.500000,
    0.750000, 1.000000, 1.500000, 2.000000, 3.000000
])

# -------------------------------
# Build interpolator
# -------------------------------
def build_local_vol_interpolator(local_vol_np, tenors, moneyness_log, F0):
    tenors = np.asarray(tenors)
    m_grid = np.asarray(moneyness_log)
    lv     = np.asarray(local_vol_np)

    def sigma(t, F):
        t_clamped = np.clip(t, tenors[0], tenors[-1])
        x = np.log(F / F0)
        x_clamped = np.clip(x, m_grid[0], m_grid[-1])

        i_hi = np.searchsorted(tenors, t_clamped)
        if i_hi == 0:
            i0, i1 = 0, 0
            wt = 0.0
        elif i_hi >= len(tenors):
            i0, i1 = len(tenors)-1, len(tenors)-1
            wt = 0.0
        else:
            i0, i1 = i_hi-1, i_hi
            t0, t1 = tenors[i0], tenors[i1]
            wt = (t_clamped-t0)/(t1-t0)

        j_hi = np.searchsorted(m_grid, x_clamped)
        if j_hi == 0:
            j0, j1 = 0, 0
            wx = 0.0
        elif j_hi >= len(m_grid):
            j0, j1 = len(m_grid)-1, len(m_grid)-1
            wx = 0.0
        else:
            j0, j1 = j_hi-1, j_hi
            x0, x1 = m_grid[j0], m_grid[j1]
            wx = (x_clamped-x0)/(x1-x0)

        s00 = lv[i0,j0]
        s01 = lv[i0,j1]
        s10 = lv[i1,j0]
        s11 = lv[i1,j1]

        s0 = s00*(1-wx) + s01*wx
        s1 = s10*(1-wx) + s11*wx
        return s0*(1-wt) + s1*wt

    return sigma

# -------------------------------
# Monte Carlo: one-step GBM
# -------------------------------
def price_call_one_step(F0, K, T, sigma_star, n_paths=2_000_000):
    rng = np.random.default_rng()
    Z = rng.standard_normal(n_paths)

    F_T = F0*np.exp(-0.5*sigma_star**2*T + sigma_star*np.sqrt(T)*Z)
    payoff = np.maximum(F_T-K,0.0)

    price = payoff.mean()
    stderr = payoff.std(ddof=1)/np.sqrt(n_paths)
    return price, stderr

# -------------------------------
# Parameters
# -------------------------------
F0 = 485.77548
T  = 0.50
moneyness = 1.40
K = moneyness*F0

# Build sigma(t,F)
sigma = build_local_vol_interpolator(local_vol_np, tenors, moneyness_log, F0)

# Look up local vol at maturity & strike
sigma_star = sigma(T, K)

# Price with 2M paths
price, err = price_call_one_step(F0, K, T, sigma_star, 200000000)

print("sigma* =", sigma_star)
print("Price:", price, "±", 2*err)
```
