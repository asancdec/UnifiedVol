```python
import sys
sys.path.insert(0, "/mnt/c/dev/UnifiedVol/python")

from unifiedvol import interp
import numpy as np

# Parametrize
S = 485.77548
r = 0.02
q = 0.00
T = 3.0
NX = 100
NT = 100

# Precompute
T_array = np.linspace(0, T, num=NT)
X_array = np.linspace(0, 3 * S, num=NX) 

X_mid = 0.5 * (X_array[:-1] + X_array[1:]) 
T_mid = 0.5 * (T_array[:-1] + T_array[1:]) 

dT = T_array[1] - T_array[0]
dX = X_array[1] - X_array[0]

lv_mid = interp.pchip_interp_2d(X_mid, T_mid, strikes, tenors, local_vol) # row time, column strike
C_mid = 0.5 * lv_mid**2 * X_mid**2 
B_mid = (r - q) * X_mid 

w = dX * B_mid / C_mid
delta = 1/w - 1/(np.exp(w) -1)

# Init condition
U_matrix = np.zeros((NT, NX))
sigma = max(dX, 0.02 * S)   # TEST            
init = np.exp(-0.5 * ((X_array - S) / sigma)**2)
init /= np.sum(init * dX)          

U_matrix[0, :] = init

#### BOTTOM IS CHATGPT CODE
## DO NOT TRUST

# ---------------------------
# Chang-Cooper fully implicit timestepping (zero-flux)
# Paste this after your existing code (does not modify your variables above)
# ---------------------------
import numpy as _np

# Parameters for stability/precision
_eps_C = 1e-14         # floor for C to avoid division by zero
_w_tol = 1e-8          # |w| < w_tol -> use Taylor series for delta
_theta = 1.0           # fully implicit

# Helper: stable delta from w (we compute our own delta_safe and will use that)
def _delta_from_w(w):
    # w: array
    small = _np.abs(w) < _w_tol
    delta = _np.empty_like(w, dtype=float)
    # cubic Taylor for small w: 1/2 - w/12 + w^3/720
    delta[small] = 0.5 - w[small] / 12.0 + (w[small] ** 3) / 720.0
    # exact (stable) for other w using expm1
    # guard against zero division by replacing exact where w==0 (shouldn't happen because small covers it)
    w_large = w[~small]
    delta[~small] = 1.0 / w_large - 1.0 / _np.expm1(w_large)
    return delta

# Ensure shapes: C_mid, B_mid, delta may be time-dependent arrays of shape (NT-1, NX-1) or 1D
# We'll accept either. Convert to arrays for safe indexing.
C_mid_a = _np.array(C_mid)
B_mid_a = _np.array(B_mid)
delta_a = _np.array(delta)   # we will overwrite with stable eval below

# If lv_mid was (NT-1, NX-1) then C_mid_a has same; else if user had 1D it stays 1D.
# We'll compute w (and delta_safe) per time slice inside the loop, using dX and our eps.
# Prepare U_matrix (you already created it, but reassign to be safe)
U = U_matrix          # alias, will fill U[n+1,:]

# time loop: there are NT time nodes, NT-1 intervals (we use mid-values indexed by n from 0..NT-2)
for n in range(0, NT - 1):
    # pick interface arrays for this time-slab
    # C/B/delta may be 2D: (NT-1, NX-1) or 1D: (NX-1,)
    if C_mid_a.ndim == 2:
        C = _np.array(C_mid_a[n, :], dtype=float)
    else:
        C = _np.array(C_mid_a, dtype=float)

    if B_mid_a.ndim == 2:
        B = _np.array(B_mid_a[n, :], dtype=float)
    else:
        B = _np.array(B_mid_a, dtype=float)

    # compute safe C and w, then delta (we ignore the earlier 'delta' var and compute fresh)
    C_safe = _np.maximum(C, _eps_C)
    w_loc = dX * B / C_safe
    delta_safe = _delta_from_w(w_loc)

    # Now compute alpha, beta on interfaces j=0..NX-2
    alpha = C_safe / dX - B * delta_safe
    beta  = -C_safe / dX - B * (1.0 - delta_safe)

    # Build L tri-diagonal entries for nodes i = 0..NX-1
    M = NX
    lower = _np.zeros(M, dtype=float)  # L[i, i-1]
    diag  = _np.zeros(M, dtype=float)  # L[i, i]
    upper = _np.zeros(M, dtype=float)  # L[i, i+1]

    # interior rows i = 1..NX-2
    for i in range(1, M-1):
        lower[i] = -alpha[i-1]
        diag[i]  = alpha[i] - beta[i-1]
        upper[i] = beta[i]

    # zero-flux BC:
    # (L U)_0  = F_{1/2} - 0 = alpha_0*U_0 + beta_0*U_1
    diag[0]  = alpha[0]
    upper[0] = beta[0]
    # (L U)_{N-1} = 0 - F_{N-3/2} = -(alpha_{N-2}*U_{N-2} + beta_{N-2}*U_{N-1})
    lower[-1] = -alpha[-1]
    diag[-1]  = -beta[-1]

    # Assemble A = I + theta*(dT/dX)*L  (we use dense tri for clarity)
    A = _np.zeros((M, M), dtype=float)
    for i in range(M):
        if i-1 >= 0:
            A[i, i-1] = ( _theta * (dT / dX) ) * lower[i]
        A[i, i] = 1.0 + ( _theta * (dT / dX) ) * diag[i]
        if i+1 < M:
            A[i, i+1] = ( _theta * (dT / dX) ) * upper[i]

    # RHS = U_n - (1-theta)*(dT/dX) * L @ U_n
    # build L matrix-vector product efficiently using tri arrays
    U_n = U[n, :].copy()
    LU = _np.zeros_like(U_n)
    # LU = L @ U_n
    for i in range(M):
        if i-1 >= 0:
            LU[i] += lower[i] * U_n[i-1]
        LU[i] += diag[i] * U_n[i]
        if i+1 < M:
            LU[i] += upper[i] * U_n[i+1]

    RHS = U_n - (1.0 - _theta) * (dT / dX) * LU

    # Solve linear system A x = RHS
    U_np1 = _np.linalg.solve(A, RHS)

    # Numerical fixes: enforce tiny negative truncation and renormalize mass for zero-flux
    # truncate small negatives due to roundoff
    U_np1[U_np1 < 0] = _np.maximum(U_np1[U_np1 < 0], -1e-16)
    # renormalize mass (should be conserved with zero-flux)
    mass_before = _np.sum(U_n * dX)
    mass_after  = _np.sum(U_np1 * dX)
    if mass_after <= 0:
        # fallback: don't renormalize if mass went non-positive (shouldn't happen)
        pass
    else:
        U_np1 *= (mass_before / mass_after)

    # store next step
    U[n+1, :] = U_np1

# put results back to U_matrix (alias already points to it)
U_matrix = U

# Diagnostics print
_mass0 = _np.sum(U_matrix[0, :] * dX)
_massN = _np.sum(U_matrix[-1, :] * dX)
print(f"[Chang-Cooper implicit] initial mass = {_mass0:.12e}, final mass = {_massN:.12e}")
print("Done time-stepping. Inspect U_matrix (shape {})".format(U_matrix.shape))

# --- price a European Call by integrating the terminal PDF ---
K = S                 # strike asked: same as S
U_T = U_matrix[-1, :] # terminal PDF (last row)
# payoff on the X grid
payoff = np.maximum(X_array - K, 0.0)

# 1) More accurate: trapezoidal integration over the X grid
price_trap = np.exp(-r * T) * np.trapz(payoff * U_T, X_array)

# 2) Simpler rectangle rule (consistent with uniform dX)
price_rect = np.exp(-r * T) * np.sum(payoff * U_T) * dX

# diagnostics: mass check
mass_trap = np.trapz(U_T, X_array)
mass_rect = np.sum(U_T) * dX

print(f"Call price (trapz) = {price_trap:.12e}")
print(f"Call price (rect)  = {price_rect:.12e}")
print(f"Terminal PDF mass (trapz) = {mass_trap:.12e}, (rect) = {mass_rect:.12e}")
```
